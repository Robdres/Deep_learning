# Properties of multiplication of matrices
For this part suppose matrices are squared with same size. We can now see that this is true with element-wise sum
$$A(B+C) = AB+AC$$
$$A(BC) = (AB)C$$

```python
A = np.array([[1,2,3],[1,2,3],[1,4,3]])
B = np.array([[1,2,3],[1,2,3],[1,2,3]])
C = np.array([[2,3,3],[4,2,3],[1,2,3]])
print( "A" )
print( A )
print( "B" )
print( B )
print( "C" )
print( C )
print( "(AB)C" )
print((A@B)@C)
print( "A(BC)" )
print(A@(B@C))
```

$$...$$
$$a_{m,1} x_1 + a_{m,2}x_2+ ... + a_{m,n}x_n = b_m$$

## Existance of the Inverse
Well, we did all of this, as for an inverse of a matrix A to exist, Ax=b most have exactly one solution for any b that we pick, meaning that if there is no solution or many then the inverse will most likely not exist (we will see what I mean in the next section), for [4,10] there is no other value that could be a result for this, you could try it but there are no other in $(\mathbb{R}^2)$. 

Well, the calculation of the inverse can be done using many techniques, but honestly they are now really interesting for me.

```python
print( "A" )
print( A )
print( "A^-1" )
print( np.linalg.inv(A) )
print( "AA^-1" )
print( np.round(A@np.linalg.inv(A)) ) # we use round for a reason we are goin to see later in the cpart of numerical computation problems
print( "A^-1A" )
print( np.round(np.linalg.inv(A)@A) )
```

So in what cases can we see if the matrix have an inverse without solving the equation. Well linear independence! When the columns of A are linearly independant then we are going to have a solution for any point inside their span. (Or column space), you can see more about this in the book, but what does this means, suppose a matrix is a family of vectors such that 
$$A = \{A_0,A_1,\dots,A_n\}$$
then we know them to be l.i. iff  for the equation
$$\alpha_0A_0 + \alpha_1A_1+\dots+\alpha_nA_n = 0$$
the only solution is when $\alpha_i = 0$, all other combinations result in a different value different than 0, let's see why I like this idea. 

Suppose we have this equation
$$ \begin{bmatrix}1\\-1\end{bmatrix}x +  \begin{bmatrix}1\\1\end{bmatrix} y= \begin{bmatrix}2\\0\end{bmatrix}$$


```python
fig,ax = plt.subplots()
ax.quiver(0, 0, 1, -1, angles='xy', scale_units='xy', scale=1, color='r')
ax.quiver(0, 0, 1, 1, angles='xy', scale_units='xy', scale=1, color='b')
ax.set_xlim([-3, 4])
ax.set_ylim([-3, 4])
plt.grid()
```

Again is easy to see the solution is x=1 and y=1, but lets try to check if the vectors are linearly independent if we try to solve this
$$ \begin{bmatrix}1\\-1\end{bmatrix}x +  \begin{bmatrix}1\\1\end{bmatrix} y= \begin{bmatrix}0\\0\end{bmatrix}$$
Same idea we are going to try to combine many times both vectors to reach 0

```python
fig,ax = plt.subplots()
a = [[0,0,1,-1,'g'],
     [1,-1,1,1,'g'],
     [2,0,1,-1,'g']]
b = [[0,0,1,1,'b'],
     [1,1,-1,1,'b'],
     [0,2,1,1,'b']]
c = [[0,0,0.5,0.5,'r'],
     [0.5,0.5,1,-1,'r'],
     [1.5,-0.5,1,-1,'r']]
d = [[0,0,-1,-1,'y'],
     [-1,-1,-1,1,'y'],
     [-2,0,1,1,'y']]
for i in [a,b,c,d]:
    for x,y,v_x,v_y,c in i:
        ax.quiver(x, y, v_x, v_y, angles='xy', scale_units='xy', scale=1, color=c)
ax.scatter(0,0) 
ax.set_xlim([-3,4])
ax.set_ylim([-3,4])
```

These are many tries that we did but little by little is it really clear that we cannot reach it [0,0] without any other combination than x=0 and y=0, for some of you the yello path can easly reach with only another vector [1,-1] and you would be correct but I want to make a parenthesis in this case. Suppose we do this 

```python
fig,ax = plt.subplots()
ax.scatter(0,0) 
ax.set_xlim([-3,2])
ax.set_ylim([-2,2])
# some may say 
d = [
    [0,0,-1,-1,'y'],
     [-1,-1,-1,1,'y'],
     [-2,0,1,1,'y'],
     [-1,1,1,-1,'y']
    ]
for i in [d]:
    for x,y,v_x,v_y,c in i:
        ax.quiver(x, y, v_x, v_y, angles='xy', scale_units='xy', scale=1, color=c)
```

We are taking 
$$  \begin{bmatrix}1\\-1\end{bmatrix}(-1) + \begin{bmatrix}1\\-1\end{bmatrix}(1) + \begin{bmatrix}1\\1\end{bmatrix}(-1) + \begin{bmatrix}1\\1\end{bmatrix}(1) = \begin{bmatrix}0\\0\end{bmatrix}$$
So after we do distribute we are going to get to this
$$   \begin{bmatrix}1\\-1\end{bmatrix}(0) +  \begin{bmatrix}1\\1\end{bmatrix}(0) = \begin{bmatrix}0\\0\end{bmatrix}$$
So yeah :D, they are l.i.

You can use many other ways to calculate it but I like this approach. Lets see a case where they are not l.i., suppose we have

# Norms 
You can relax now, we passed linear independence. Now Another topic :D  Norms, this is usually more complicated but we can stay with the idea that norms are a way to measure a vector they have to suffice this
$$f(x) \implies x=0 $$
$$f(x+y) \leq f(x)+f(y) $$
$$\forall \alpha \in \mathbb{R}, f(\alpha x )  = |\alpha|x $$

thus Euclidean norm (Pythagoras thorem for R2 and also $L^2$), max value, etc, are norms. In machine learning we are going to measure with the $L^p$ norm define as
$$||x||_p = \left( \sum_{i} |x_i|^p \right)^{\frac{1}{p}}$$

$L^1$ norm  is
$$|x|_1  = \sum_i|x_i|$$
$L^2$ norm is
$$|x|_2  = \sqrt{\sum_i(x_i)^2}$$
$L^\infty $  also called the max norm can be define as such 
$$|x|_\infty  = \max_i|x_i|$$
That was kinda boring lets use some graphs to understand it, we are normally familiar with the $L^2$ norm the euclidean norm, for example, when we define a circle in R2, we usually tend to get a radius and a center, a circle in R2, is just all the points that are separated by the radius from the center, suppose in all the future examples that the center is [0,0], and first lets draw a circle

```python
fig,ax = plt.subplots()
ax.add_patch( plt.Circle( (0,0), 1,fill=False))
ax.set_xlim([-2, 2])
ax.set_ylim([-2, 2])
ax.set_aspect('equal')
ax.scatter(0,0,color="black")
```

So okay cool, now lets try to see what a circle is for another norm using previous definition, a set of points that have the same distance from the origin [0,0], in this case the distance or radius will be 1

### Diagonal Matrices
A diagonal matrix is the one where all elements that are not in the diagonal is 0, then 
$$D = \begin{bmatrix} a_0 & 0 & 0 & ...\\0 & a_1 & 0 & ...\\... & ... & ... & ...\\0 & 0 & ... & a_n \end{bmatrix}$$
We are going to use an specific way to define this matrix, using a vector, where all elements in the diagonal are the elements of the vector, thus
$$D = diag([a_0,a_1,a_2,...,a_n])$$
Also with this Matrix we can see some things as
- $D^{-1} $ exists if all elements are non zero
- If $D^{-1} $ exists  then $D^{-1} = diag([\frac{1}{a_0},\frac{1}{a_1},...,\frac{1}{a_n}] )$
- when multiplying this matrix by a vector we are just scaling each value then

(They can also be non squared but... you get the point)

```python
a = np.array([1,2,3,4,5,6])
b = np.array([2,2,2,2,2,2])
print("a")
print(a)
print("diag(a)")
print(np.diag(a))
print("diag(a)^{-1}")
print(np.linalg.inv(np.diag(a)))
print("b")
print(b)
print("diag(a)*b")
print(np.diag(a)@b)
```

# Eigenvalues and Eigen vectors
An eigenvector of a square matrix is a vector that is only scaled when we apply the matrix to it. What does that mean? well... 
$$Av = \lambda v$$
That means that the vector does not change direction, the $\lambda$ associated to it is know as eigenvalue, this is easier with examples.

Lets start easy, with the identity matrix let
$$I = \begin{bmatrix}1&0\\0&1\end{bmatrix}$$
Now for any v=[a,b] we have that
$$Iv = 1*v$$
it does not change, thus in this case the any vector in the R2 plane is an eigenvector and as you can see the only eigenvalue is 1. (Every eigenvector is associated a 1 as an eigenvalue, but we consider to be the same eigen value)

## Next step 
Now lets consider the matrix $\alpha I$, is easy to see that
$$\forall v \in \mathbb{R}^2 \implies \alpha I * v^T = \alpha v$$ 
Is the same, all vectors are eigenvectors but the only eigenvalue is $\alpha$, we can check out what the transformation does to a vector and a unit circle (circle of radious one). And the following matrix
$$\begin{bmatrix}2&0\\0&2\end{bmatrix}$$

Small editor's note: In python 2D's transformations can be handle as below, but we handle this 2D transformations with a 3D matrix of this form
$$\begin{bmatrix}a&b&0\\c&d&0\\0&0&1\end{bmatrix}$$
You can see the explanation of why is done like that here [Link](https://medium.com/@junfeng142857/affine-transformation-why-3d-matrix-for-a-2d-transformation-8922b08bce75)

```python
import matplotlib.transforms as tra
fig, ax = plt.subplots(ncols = 2)
b = np.array([[2,0,0],[0,2,0],[0,0,1]])
print(b)
ax1 = ax[0]
ax2 = ax[1]
x = np.linspace(-3,3, 100)
y = np.linspace(-3,3, 100)

t = tra.Affine2D(b) + ax2.transData

circle1 = plt.Circle([0,0],1,fill=False)

ax1.quiver(0, 0, 1, 0, angles='xy', scale_units='xy', scale=1, color='k')
ax1.quiver(0, 0, 0, 1, angles='xy', scale_units='xy', scale=1, color='k')
ax1.add_patch(circle1)

circle = plt.Circle([0,0],1,fill=False,transform=t,color ="r")
ax2.quiver(0, 0, 2, 0, angles='xy', scale_units='xy', scale=1, color='r')
ax2.quiver(0, 0, 0, 2, angles='xy', scale_units='xy', scale=1, color='r')
ax2.add_patch(circle)

ax1.axis('square')
ax1.set_title("v")
ax2.axis('square')
ax2.set_title("A*v")
ax1.set_ylim(-3, 3)
ax1.set_xlim(-3, 3)
ax2.set_ylim(-3, 3)
ax2.set_xlim(-3, 3)

plt.show()
```

# Eigen Matrix descomposition

Where Q is the matrix of eigenvectors and $\lambda$'s are the eigen values


```python
A = np.array([[2,1],[1,1]])
print("A")
print(A)
print("U*D*U.T")
print(np.linalg.eig(A)[1] @ np.diag(np.linalg.eigvals(A)) @np.linalg.eig(A)[1].T)
```

# Psedudo Inverse
Well until now we have work in square matrices, as those are the only ones that can have an inverse, but is necessary to understand what could be the equivalent to the inverse for non-square matrices. This pseudo-inverse is define as 
$$A^+ = \lim_{\alpha->0} (A^TA + \alpha I )^{-1}A^T$$
There is some cases when the matrix is taller than wide (meaning less vectors than necessary) then the solution $x=A^xb$ is going to give one of the many solutions, and if is more wider than tall, it may not have a solution but it could give the closer point to the solution, all in terms to the $L^2$ norm

```python
A = np.array([[2,3],[1,2],[3,4]])
print("A")
print(A)
print("A^+")
print(np.linalg.pinv(A))
B = np.array([[2,3,3],[1,2,3]])
print("B")
print(B)
print("B^+")
print(np.linalg.pinv(B))
```
